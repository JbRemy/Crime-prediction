```python
df['Offense_cat'] = df.Text_General_Code

#Cambriolages
df.Offense_cat = df.Offense_cat.replace("Burglary Non-Residential", "Burglary").copy()
df.Offense_cat = df.Offense_cat.replace("Burglary Residential", "Burglary").copy()

#Vols
df.Offense_cat = df.Offense_cat.replace("Robbery Firearm", "Robbery").copy()
df.Offense_cat = df.Offense_cat.replace("Robbery No Firearm", "Robbery").copy()

#Homicides
df.Offense_cat = df.Offense_cat.replace("Homicide - Criminal", "Homicide").copy()
df.Offense_cat = df.Offense_cat.replace("Homicide - Gross Negligence", "Homicide").copy()
df.Offense_cat = df.Offense_cat.replace("Homicide - Justifiable", "Homicide").copy()

#arnaques
df.Offense_cat = df.Offense_cat.replace("Fraud", "Fraud").copy()
df.Offense_cat = df.Offense_cat.replace("Gambling Violations", "Fraud").copy()
df.Offense_cat = df.Offense_cat.replace("Receiving Stolen Property", "Fraud").copy()
df.Offense_cat = df.Offense_cat.replace("Recovered Stolen Motor Vehicle", "Fraud").copy()
df.Offense_cat = df.Offense_cat.replace("Embezzlement", "Fraud").copy()
df.Offense_cat = df.Offense_cat.replace("Forgery and Counterfeiting", "Fraud").copy()

#Vols
df.Offense_cat = df.Offense_cat.replace("Thefts", "Theft").copy()
df.Offense_cat = df.Offense_cat.replace("Motor Vehicle Theft", "Theft").copy()
df.Offense_cat = df.Offense_cat.replace("Theft from Vehicle", "Theft").copy()

#order
df.Offense_cat = df.Offense_cat.replace("Disorderly Conduct", "Order").copy()
df.Offense_cat = df.Offense_cat.replace("DRIVING UNDER THE INFLUENCE", "Order").copy()
df.Offense_cat = df.Offense_cat.replace("Public Drunkenness", "Order").copy()
df.Offense_cat = df.Offense_cat.replace("Public Drunkenness", "Order").copy()
df.Offense_cat = df.Offense_cat.replace("Prostitution and Commercialized Vice", "Order").copy()
df.Offense_cat = df.Offense_cat.replace("Narcotic / Drug Law Violations", "Order").copy()
df.Offense_cat = df.Offense_cat.replace("Liquor Law Violations", "Order").copy()


#Aggressions physiques
df.Offense_cat = df.Offense_cat.replace("Offenses Against Family and Children", "Assault").copy()
df.Offense_cat = df.Offense_cat.replace("Vandalism/Criminal Mischief", "Assault").copy()
df.Offense_cat = df.Offense_cat.replace("Aggravated Assault Firearm", "Assault").copy()
df.Offense_cat = df.Offense_cat.replace("Aggravated Assault No Firearm", "Assault").copy()
df.Offense_cat = df.Offense_cat.replace("Weapon Violations", "Assault").copy()
```

```python
import datetime
```

```python
def new_date(x):
    date = datetime.date(int(x[0:4]),int(x[5:7]),int(x[8:10]))
    Year = date.strftime("%Y")
    Month = date.strftime("%B")[0:3]
    Day = date.day
    Nb_month_since_2006 = (int(x[0:4])-2006)*12 + int(x[5:7])-1
    return([Year,Month,Day,Nb_month_since_2006])

df_temp = pd.DataFrame(list(map(new_date, np.array(df.Dispatch_Date))),columns=['Year','Month','Day','Month_since_2006'])
```

```python
df = pd.concat([df_temp,df],axis=1)
```

### Creation de la Time serie

```python
df.Offense_cat.unique()
```

```python
Time_serie = pd.DataFrame(columns = ['Other Assaults', 'All Other Offenses', 'Assault', 'Theft',
       'Burglary', 'Order', 'Arson', 'Fraud', 'Robbery',
       'Vagrancy/Loitering'], index = range(0,132))

for i in df[df.Offense_cat.isnull() == False].Offense_cat.unique():
    Time_serie[i] = df.groupby('Offense_cat').get_group(i).groupby('Month_since_2006').size()
```

```python
Time_serie['Crimes'] = df.groupby('Month_since_2006').size()
```

```python
Time_serie
```

```python
for i in Time_serie.columns :
    if Time_serie.isnull()[i].sum() != 0:
        print(i)
```

```python
Time_serie['Arson'][Time_serie['Arson'].isnull()==True] = 0
Time_serie['Fraud'][Time_serie['Fraud'].isnull()==True]  = 0
Time_serie['Vagrancy/Loitering'][Time_serie['Vagrancy/Loitering'].isnull()==True]  = 0
Time_serie['Rape'][Time_serie['Rape'].isnull()==True]  = 0
Time_serie['Other Sex Offenses (Not Commercialized)'][Time_serie['Other Sex Offenses (Not Commercialized)'].isnull()==True]  = 0
Time_serie['Homicide'][Time_serie['Homicide'].isnull()==True]  = 0
```

```python
Time_serie = Time_serie.astype(int)
```
