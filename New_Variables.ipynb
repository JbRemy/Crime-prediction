### Mise en forme sous un seul Data Frame

On veut désormais regrouper ces données en un seul Data Frame, où une ligne correspond à un mois (à partir du 1er Janvier 2006).

```python
df_temp = pd.DataFrame(columns=["Year", "Month", var], index=(range(0,132)))
```

```python
a=list(df1.columns)
a.remove("Year")
var = ["Mining_Logging_Construction", "Manufacturing", "Trade_Transportation_Utilities", "Information",
      "Financial_Activities", "Professional_and_Business_Services", "Education_and_Health_Services",
      "Leisure_and_Hospitality", "Other_Services", "Government"]
dfs = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10]
dfs_2 = []

for i,var in zip(dfs,var):
    df_temp = pd.DataFrame(columns=[var], index=(range(0,132)))
    j=0
    for y in range(0,11):
        for Month in a:
            df_temp[var].ix[j] = i[Month].ix[y]
            j+=1
    dfs_2.append(df_temp)
```

```python
dfs_2.append(df)
data = pd.concat(dfs_2,axis=1)
```
